## Loading a dataset

```python
import pandas as pd
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'label'])
iris
```

```python
X = iris.drop(['label'], axis=1)
X
```

## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
model = PCA(n_components=4)  # n_components specifies the new dimensionality
X_reduced = model.fit(X).transform(X) # or just pca.fit_transform(X)
pd.DataFrame(X_reduced)
```

다른 숫자들이 나옴. 축이 이동하고 난 상태의 값들

출력값을 X_reduced로 두고 dataframe으로 보기 좋게 정리한 것

```python
print('Components:', model.components_)
print('Explained Variance:', model.explained_variance_)
print('Explained Variance Ratio:', model.explained_variance_ratio_ )
```

model.components를 출력한 정보. 행렬
pca한 값을 원상복구 해줄 수 있는 정보. pca 후의 값에 이 정보들을 곱하면 원래 값으로 복구

150X4, 4X4 -> 150X4 로 복구

(행렬은 직사각형이다.

3x2  x  2x5 -> 3x5 행렬이 나옴. )

variance는 sd의 제곱. 클수록 더 중요한 정보. 설명력, 두께

n_components=2로 하면 원복하는 행렬이 다르게 나옴. 2X4로 나옴. 

```python
X_approx = model.inverse_transform(X_reduced) 
pd.DataFrame(X_approx)
```

pca후의 행렬을 넣어서 x_approx

오리지널이나올 수 없음. 2개의 component만 선택했다면, 원래의 자료가 완전히 나오지 않음. 

이렇게 구한 값이랑 원래 자료랑 비교하면, 차이가 있긴 하지만 거의 비슷한 값이 나오는 것을 볼 수 있음. 

1개만 갖고 해도 전체의 대부분을 차지해서 이정도로 차원축소해도 가능함. 

pca 된 값들은 서로 decorrelate 된 값이다.
